```python
# Import libraries and load the dataset

import pandas as pd
import numpy as np

# Load the dataset from the provided URL
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")

# Display the first 10 rows
df.head(10)
```

```python
# Check for missing values

# Calculate percentage of missing values in each column
df.isnull().sum() / len(df) * 100
```

```python
# Identify numerical and categorical columns

# Check the data types of each column
df.dtypes
```

```python
# Count the number of launches at each launch site

# Count the number of launches for each site
df['LaunchSite'].value_counts()
```

```python
# Count the number and occurrence of each orbit type

# Count the number of occurrences for each orbit type
df['Orbit'].value_counts()
```

```python
#Count the number and occurrence of mission outcomes

# Count the number of each mission outcome
landing_outcomes = df['Outcome'].value_counts()

# Display the result
landing_outcomes
```

```python
# Print each outcome with its index number
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)
```

```python
# Create a landing outcome label (0 = failure, 1 = success)

bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
# Create a new column 'Class' based on whether the outcome is in the bad_outcomes set
landing_class = []

for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

df['Class'] = landing_class

# Show the first few rows of the new column
df[['Outcome', 'Class']].head(8)
```
